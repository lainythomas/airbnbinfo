```python
import pandas as pd
import seaborn as sns
listings = pd.read_csv('listings.csv')
listings.head()
```

```python
listings.describe()
```

```python
listings['neighbourhood_group'].unique()
```

```python
totals = listings['neighbourhood'].value_counts()
top20popularcities = totals.head(20)
top20popularcities
```

```python
listings['room_type'].unique()
```

```python
sns.countplot(data = listings, y = "room_type")
```

```python
top20cities = listings[listings['neighbourhood'].isin(top20popularcities.index)]
top20cities
```

```python
avgprice = top20cities.groupby("neighbourhood").agg({"price":"mean"})
maxprice = top20cities.groupby("neighbourhood").agg({"price":"max"})
minprice = top20cities.groupby("neighbourhood").agg({"price":"min"})
avgprice
#We can see that Hollywood Hills West has the most expensive Airbnbs on average, followed by Beverly Hills, 
#Beverly Grove, Hollywood Hills, Venice.

#Hollywood Hills West and Beverly Hills are sufficiently more expensive than the other top 20 most visited cities.
```

```python
from matplotlib import pyplot as plt

top5 = ['Hollywood Hills West', 'Beverly Hills', 'Beverly Grove', 'Hollywood Hills', 'Venice']
top5cities = top20cities.loc[top20cities['neighbourhood'].isin(top5)]
plt.figure(figsize = (15,10))
sns.boxplot(data = top5cities,
           x = 'neighbourhood',
           y = 'price')

#Beverly Hills and Hollywood Hills West have the most variability in price, while the next 3 cities have a
#smaller price range with the max around $4000.
```

```python
sns.scatterplot(data = top20cities,
             x = 'price',
             y = 'number_of_reviews')

#The lower the price, the more reviews the Airbnb property has. This makes sense because more people can afford
#properties at a lower price point and therefore there will likely be more people staying at these properties.
#Additionally, Airbnbs are usually booked for short periods of time, so spending less money is ideal.
```

```python
top20cities['minimum_nights'].value_counts()
#Most Airbnbs have a minimum stay of 30 nights. This is interesting because Airbnb started out as more of a 
#short-term service. Now, there are many more people looking to lease their property for at least a month at
#a time -- 6389 out of 13965 properties within the top 20 most popular cities (about 46%).
```

```python
heatmapdata = pd.crosstab(top5cities['room_type'], top5cities['neighbourhood'])
heatmapdata
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(heatmapdata, annot = True)

#There are 1277 entire homes/apartments listed on Airbnb in Venice, which is much more than the other 4 cities
#more expensive than Venice (Hollywood Hills West, Beverly Hills, Beverly Grove, and Hollywood Hills - 382, 
#362, 315, and 573 respectively). Venice has the most options for properties in terms of quantity on Airbnb.
```
